```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Faulty_dataset.csv')
df
```

```python
def fillNames(row):
    if pd.isnull(row['name']):
        if row['gender']=="Male":
            return "John Doe"
        elif row['gender']=='Female':
            return "Jane Doe"
    return row['name']

df['name']=df.apply(fillNames,axis=1)
df
```

```python
column_name = '10th'  # Replace with the actual column name
lower_limit = 0
upper_limit = 100

# Remove rows where the column value is outside the specified range
df = df[(df[column_name] >= lower_limit) & (df[column_name] <= upper_limit)]
df
```

```python
column_name = '10th'  # Replace with the actual column name
lower_limit = 0
upper_limit = 100
replace1=100
replace2=0
df.loc[(df[column_name] < lower_limit), column_name] = replace1
df.loc[ (df[column_name] > upper_limit),column_name]=replace2
df
```

```python
lb=0
ub=100
df['10th']=df['10th'].apply(lambda x:lb if x<lb else(ub if x>ub else x))
df
```
