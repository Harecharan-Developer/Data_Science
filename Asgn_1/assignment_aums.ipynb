```python
df3["actual_year"]=pd.to_datetime(df3["Year"],format='%m/%d/%y %H:%M')
df3["YEAR"]=df3["actual_year"].dt.year
df3.drop(columns="actual_year",inplace=True)
df3
```

```python
df3.groupby("Educational Attainment")["Population Count"].sum()
```

2

```python
df3[(df3["Age"] != "00 to 17")].groupby("Educational Attainment")[["Population Count"]].sum()/df3.groupby("Educational Attainment")[["Population Count"]].sum()*100
```

```python
percentage_distribution = (df3.query("Age != '00 to 17'").groupby("Educational Attainment")[["Population Count"]].sum() / df3.groupby("Educational Attainment")[["Population Count"]].sum()) * 100
percentage_distribution
```

```python
df3.info()
```

```python
df3[["l_age","u_age"]]=df3["Age"].str.extract(r'(\d+) to (\d+|\d+\+)')
df3
```

3

```python
contigencyTable = pd.pivot_table(df3,index='Educational Attainment',columns='Personal Income',aggfunc=len,fill_value=0)
contigencyTable
```

4

```python
no_diploma = df3[df3["Educational Attainment"] == "No high school diploma"]
no_diploma = no_diploma.groupby(by="Personal Income")["Population Count"].sum()

bach_higher = df3[df3["Educational Attainment"] == "Bachelor's degree or higher"]
bach_higher = bach_higher.groupby(by="Personal Income")["Population Count"].sum()

x_plot_value = no_diploma.index.to_numpy()
x_plot_value2 = bach_higher.index.to_numpy()

y_plot_value = no_diploma.values
y_plot_value2 = bach_higher.values
```

```python
x_axis = np.arange(len(x_plot_value))
width = 0.3
plt.barh(x_axis, y_plot_value, width, color='yellow', label="No Diploma")
plt.barh(x_axis + 0.3, y_plot_value2, width, label="Bachelor's or Higher")
plt.yticks(x_axis, x_plot_value)
plt.ylabel("Frequency")
plt.xlabel("Personal Income")
plt.legend()
plt.show()
```
