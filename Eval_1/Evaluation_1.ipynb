```python
#q12
age_of_men = df3[df3['sex'] == ' Male']['age']

# Create a histogram with bins of size 20
plt.hist(age_of_men, bins=np.arange(0, age_of_men.max() + 20, 20), edgecolor='k', alpha=0.7, color='skyblue')

# Add labels and a title
plt.xlabel('age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram of Age for Men', fontsize=16)

# Calculate the skewness
skewness = age_of_men.skew()

# Comment on skewness
if skewness > 0:
    comment = 'positive skew'
elif skewness < 0:
    comment = 'negative skew'
else:
    comment = 'approximately symmetric'

print(f'Skewness of Age Distribution for Men: {skewness:.2f} ({comment})')

# Show the plot
```
